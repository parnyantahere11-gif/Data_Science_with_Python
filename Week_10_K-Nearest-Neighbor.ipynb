```python
### K Nearest Neighbors ### 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python

## Get the data

df= pd.read_csv('/Users/taherehehsan/Classified Data.txt', index_col=0)
```

```python
df.head()
```

```python

## Preprocessing

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
#fit the mean and sd per column

scaler.fit(df.drop('TARGET CLASS', axis=1))
```

```python
# transform the data to scaled features

scaled_features= scaler.transform(df.drop('TARGET CLASS', axis= 1))
```

```python
scaled_features
```

```python
# turn the scaled features from array into Dataframe

df_feat = pd.DataFrame(scaled_features, columns = df.columns[:-1])
df_feat.head()
```

```python


## Train Test Split

from sklearn.model_selection import train_test_split
```

```python
 x_train, x_test, y_train, y_test = train_test_split(scaled_features, df['TARGET CLASS'],
                                                    test_size=0.30)
```

```python


## KNeighborsClassifier & Prediction

from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors= 1)
```

```python
knn.fit(x_train, y_train)
```

```python
pred = knn.predict(x_test)
```

```python

#### KNN 
## Evaluation

from sklearn.metrics import classification_report, confusion_matrix
```

```python

print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

```python

#### KNN 
## Elbow method

error_rate= []

for i in range(1,40):

   knn = KNeighborsClassifier(n_neighbors= i)
   knn.fit(x_train, y_train)
   pred_i = knn.predict(x_test)
   error_rate.append(np.mean(y_test != pred_i))
```

```python
error_rate
```

```python

## Elbow method figure

plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
# FIRST A QUICK comparison TO OUR ORIGINAL K=1 

knn = KNeighborsClassifier(n_neighbors= 1)
   
knn.fit(x_train, y_train)
pred = knn.predict(x_test)

print('WITH K=1')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

```python
# NOW WITH K=15

knn = KNeighborsClassifier(n_neighbors= 17)
   
knn.fit(x_train, y_train)
pred = knn.predict(x_test)

print('WITH K=17')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

```python
# AND WITH K=30

knn = KNeighborsClassifier(n_neighbors= 30)
   
knn.fit(x_train, y_train)
pred = knn.predict(x_test)

print('WITH K=30')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

```python

## Get the Data

df = pd.read_csv('/Users/taherehehsan/Classified Data.txt', index_col = 0)
```

```python

## Standardize the Variables

from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
scaler.fit(df.drop('TARGET CLASS', axis= 1))
```

```python
scaled_features = scaler.transform(df.drop('TARGET CLASS', axis= 1))
```

```python
df_feat = pd.DataFrame(scaled_features, columns =df.columns[:-1])
df_feat
```

```python
df_feat.head()
```

```python

## Train Test Split

from sklearn.neighbors import KNeighborsClassifier
```

```python

x_train, x_test, y_train, y_test = train_test_split(scaled_features, df['TARGET CLASS'],
                                                   test_size = 0.30)
```

```python

## Using KNN

from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors= 1)
```

```python
knn.fit(x_train, y_train)
```

```python
pred = knn.predict(x_test)
```

```python

## valuations

from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

```python

### Choosing a K Value

error_rate = []

#will take some time
for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python

plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
# FIRST A QUICK COMPARISON TO OUR ORIGINAL K=1
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(x_train,y_train)
pred = knn.predict(x_test)

print('WITH K=1')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

```python

# NOW WITH K=15
knn = KNeighborsClassifier(n_neighbors=17)

knn.fit(x_train,y_train)
pred = knn.predict(x_test)

print('WITH K=15')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
