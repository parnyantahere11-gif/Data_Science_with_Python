```python

#######################

#################

##########

#### K-Means Clustering

import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
## K-Means on a rndomly generated dataset
np.random.seed(0)
```

```python
x, y= make_blobs(n_samples= 5000, centers=[[4,4], [-2, -1], [2, -3], [1, 1]], cluster_std= 0.9 )
```

```python
## Display of the randomly generated data

plt.scatter(x[:,0], x[:,1], marker='.')
```

```python
x
```

```python
type(x)
```

```python
y
```

```python

## Setting up K-Means

# k-means++: Selects initial cluster centers for 
# k-mean clustering in a smart way to speed up convergence.

# n_init: Number of time the k-means algorithm will be run with different centroid seeds
# The final results will be the best output of n_init consecutive runs in terms of inertia.

k_means = KMeans(init = 'k-means++', n_clusters = 4, n_init= 12)
```

```python
k_means.fit(x)
```

```python
k_means_labels = k_means.labels_
k_means_labels
```

```python
k_means_cluster_centers = k_means.cluster_centers_
k_means_cluster_centers
```

```python

## Creating the Visual Plot

fig = plt.figure(figsize= (6,4))

colors = plt.cm.Spectral(np.linspace(0, 1, len(set(k_means_labels))))

ax = fig.add_subplot(1, 1, 1)

for k, col in zip(range(len([[4,4], [-2, -1], [2, -3], [1, 1]])), colors):
    my_members = (k_means_labels == k)
    cluster_center = k_means_cluster_centers[k]
    ax.plot(x[my_members, 0], x[my_members, 1], 'w', markerfacecolor=col, marker='.')
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,  markeredgecolor='k', markersize=6)

ax.set_title('KMeans')

# Remove x-axis ticks
ax.set_xticks(())

# Remove y-axis ticks
ax.set_yticks(())

# Show the plot
plt.show()
```

```python
# kMeans = 3


k_means3 = KMeans(init = 'k-means++', n_clusters = 3, n_init= 12)
k_means3.fit(x)
```

```python
fig = plt.figure(figsize= (6,4))

colors = plt.cm.Spectral(np.linspace(0, 1, len(set(k_means3.labels_))))

ax = fig.add_subplot(1, 1, 1)

for k, col in zip(range(len(k_means3.cluster_centers_)), colors):
    my_members = (k_means3.labels_ == k)
    cluster_center = k_means3.cluster_centers_[k]
    ax.plot(x[my_members, 0], x[my_members, 1], 'w', markerfacecolor=col, marker='.')
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,  markeredgecolor='k', markersize=6)

```

```python

## Customer Segmentation with K_means

import pandas as pd
```

```python
cust_df = pd.read_csv('/Users/taherehehsan/Cust_Segmentation.csv')
```

```python
cust_df.head(5)
```

```python
## Preprocessing 

df = cust_df.drop('Address', axis=1)
```

```python
df.head()
```

```python
## Normalizing the data because we want to apply mathematical_based algorithm to 
# interpret features with different magnitude and distribution equally.

from sklearn.preprocessing import StandardScaler
```

```python
# [all rows, all columns except o and 1(index and ID) ]
X = df.values[:, 1:]
```

```python
# change all nan to a number = 0
X= np.nan_to_num(X)
```

```python
# scales the data so each feature has mean = 0 and standard deviation = 1 (normalization)
clust_dataset = StandardScaler().fit_transform(X)
clust_dataset
```

```python

## Modeling
clusterNum = 3
k_means = KMeans(init= 'k-means++', n_clusters= clusterNum, n_init = 12)
```

```python
k_means.fit(X)
labels= k_means.labels_
print(labels)
```

```python
df['clus_km']= labels
```

```python
df.head(5)
```

```python
## check the centroid values by averaging the features in each cluster
# ه هر سطر یک شماره خوشه می‌دهد.
# میانگین ویژگی‌ها برای هر خوشه حساب میشه.
# .هستند (centroid) این میانگین‌ها تقریباً همون مرکز خوشه 

df.groupby('clus_km').mean()
```

```python
# Check the distribution of customers

area = np.pi * (X[:, 1])**2  

# Scatter plot (replace np.float with float)
plt.scatter(X[:, 0], X[:, 3], 
            s=area, 
            c=labels.astype(float), 
            alpha=0.5)

plt.xlabel('Age', fontsize=18)
plt.ylabel('Income', fontsize=16)
plt.show()

```

```python
## Or
area = np.pi * (df['DebtIncomeRatio'])**2  

plt.scatter(df['Age'], df['Income'], 
            s=area, 
            c=df['clus_km'], 
            cmap='tab10', 
            alpha=0.5)

plt.xlabel('Age', fontsize=14)
plt.ylabel('Income', fontsize=14)
plt.title('Customer Segmentation: Age vs Income', fontsize=16)
plt.show()
```
