```python

## Multi-Index and Index Hierarchy ##

# Index Levels

outside = ['G1','G1','G1','G2','G2','G2']
```

```python
inside = [1,2,3,1,2,3]
```

```python
hier_index = list(zip(outside,inside))
```

```python
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
hier_index
```

```python
f = pd.DataFrame(np.random.randn(6,2),index=hier_index,columns=['A','B'])
```

```python
f
```

```python
## or 

outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]

hier_index = pd.MultiIndex.from_arrays([outside, inside])

f = pd.DataFrame(np.random.randn(6,2), index=hier_index, columns=['A','B'])
```
