```python

## PCA

from sklearn.decomposition import PCA
```

```python
x_pca.shape
```

```python

## Visualising

plt.figure(figsize= (8,6))

plt.scatter(x_pca[:,0], x_pca[:,1], c= cancer['target'], cmap= 'plasma' )

plt.xlabel('First principal component')

plt.ylabel('Second principal component')
```

```python

## With Seaborn

sns.scatterplot(x = x_pca[:,0], y= x_pca[:,1], hue= cancer['target'],palette= 'rainbow' )

plt.xlabel('First principal component')

plt.ylabel('Second principal component')
```

```python
## what is the share of each feature in influence on the Component 

pca.components_
```

```python
pca.components_.shape
```

```python

## change this array with 2 rows and 30 columns to a DataFrame

df_comp = pd.DataFrame(pca.components_, columns= cancer['feature_names'])
df_comp
```

```python

## visual the share of each feature by heatmap

plt.figure(figsize=(12,6))

sns.heatmap(df_comp, cmap= 'plasma')
```

```python

## The Data

from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
type(cancer)
```

```python
cancer
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
# Turn this dict to DF

df = pd.DataFrame(cancer['data'], columns= cancer['feature_names'])
```

```python
df.head()
```

```python
df.info()
```

```python
#### PCA Visualization

## Scaling Data

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_data = scaler.fit_transform(df)
```

```python
scaled_data
```

```python
type(scaled_data)
```

```python
type(scaler)
```

```python
scaled_data.shape
```

```python

## PCA Applying

from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
x_pca = pca.fit_transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python

### DONE
```

```python

## Visualizing with matplotlib

plt.figure(figsize= (8,6))

plt.scatter(x_pca[:,0], x_pca[:, 1], c = cancer['target'], cmap= 'plasma')

plt.xlabel('First Principal Component')

plt.ylabel('Second Principal Componant')
```

```python

## Visualizing with matplotlib

sns.scatterplot(x= x_pca[:,0], y=x_pca[:,1], hue= cancer['target'], palette= 'rainbow')

plt.xlabel('First Principal Component')

plt.ylabel('Second Principal Componant')
```

```python
## What is the difference between these two figures?? Think about it!
```

```python
df_feat = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df_target = pd.DataFrame(cancer['target'],columns=['Cancer'])
df1 = pd.concat([df_feat, df_target], axis=1)
sns.scatterplot(x='mean radius',y='mean texture',hue='Cancer',data=df1)
```

```python
## first plot is showing the whole features which try to plot the trend per each component,
## in second one we have 2 features which may lead the component we have or in the other word:
# they have a higher share to influence first and second component respectively
## because they have almost same trend
```

```python

## Interpreting the components

pca.components_
```

```python
pca.components_.shape
```

```python

# Turn this array to a DataFrame to see the share of each feature on each component

df_comp= pd.DataFrame(pca.components_, columns= cancer['feature_names'])
```

```python
df_comp
```

```python
## Using heatmap to see their influence 

plt.figure(figsize=(12,6))

sns.heatmap(df_comp, cmap = 'plasma')
```

```python
## This heatmap and the color bar basically represent the correlation between
# the various feature and the principal component itself.
```
