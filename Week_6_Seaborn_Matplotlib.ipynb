```python
import seaborn as sns
```

```python

####  Seaborn Library

## Built-In Dataset

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python


##### countplot() for plotting bar chart

# Bar chart 1

sns.countplot(x = 'sex', data= tips, color= 'red')
```

```python

## bar chart 2 
sns.countplot(x= 'day',data= tips )
```

```python

## Side by side bar chart by hue

sns.countplot(x= 'sex' , data = tips, hue= 'day')
```

```python

## Histogram by distplot()  for Numarical DataSet

x = tips['total_bill']

sns.distplot(x)

# safe to ignore warnings
```

```python

## By default kde= TRUE that shows a statistical curve but When we want to remove that 
## We should write kde=FALSE

sns.distplot(x, kde = False )
```

```python

### scatterplot()
## It is used when we want to plot 2 numerical parameter

sns.scatterplot(x= tips['total_bill'], y= tips['tip'] )
```

```python

### boxplot()
## When we have both numerical and categorical to plot

sns.boxplot(x= 'day', y= 'total_bill', hue='day', data = tips, palette = 'rainbow')
```

```python

## pairplot

sns.pairplot(tips)
```

```python

## pairplot( hue= )

sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

```python

## jointplot()
# how two parameter are related to each other: scatterplot + histogram

sns.jointplot(x = 'total_bill', y= 'tip', data= tips,kind='scatter' )
```

```python

## jointplot(kind)

## scatterplot + histogram + regression

sns.jointplot(x = 'total_bill', y= 'tip', data= tips, kind= 'reg')
```

```python

## Matplotlib has subplot 
## seaborn does not give us this facilities easily

# plt.subplot(nrows, ncols, plot_number)

plt.subplot(1,2,1)
sns.countplot(x= 'sex', data= tips, hue='sex', palette= 'rainbow')
plt.subplot(1,2,2)
sns.countplot(x= 'day',  hue='day', data= tips, palette= 'rainbow')
```

```python
%matplotlib inline
```

```python

## Data

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python

## Counplot : bar chart for categorical data

sns.countplot(x= 'sex', data= tips)
```

```python
sns.countplot(x ='day', data = tips, color = 'red')
```

```python

## Countplot(hue= ) : side_by_side bar chart for categorical data

sns.countplot(x= 'sex', data= tips, hue= 'day')
```

```python

## Distplot  : histogram for numerical data

x = tips['total_bill']
```

```python
sns.displot(x)
```

```python
sns.displot(x, kde= False)
```

```python

## Scatterplot: for showing two numerical parameters

sns.scatterplot(x=tips['total_bill'], y= tips['tip'])
```

```python

## Boxplot: for showing both numerical and categorical data

sns.boxplot(x= 'day', y= 'total_bill', hue = 'day', data= tips, palette= 'rainbow')
```

```python

## pairplot: for having idea about the whole data at the begining

sns.pairplot(tips)
```

```python

sns.pairplot(tips,hue='sex', palette= 'coolwarm')
```

```python

## jointplot(kind= 'scatter') : the relation between 2 parameters by Scatterplot + Histogram

sns.jointplot(x='total_bill', y= 'tip', data= tips, kind= 'scatter')
```

```python

## jointplot(kind= 'reg') : the relation between 2 parameters by Scatterplot + Histogram + regression 

sns.jointplot(x= 'total_bill', y='tip', data= tips, kind= 'reg')
```

```python

## We can combine matplotlib with seaborn
#plt.subplot(nrows, ncols, plot_number )

plt.subplot(1,2,1)
sns.countplot(x= 'sex', hue= 'sex', data=tips, palette= 'rainbow')
plt.subplot(1,2,2)
sns.countplot(x= 'day', hue= 'day', data= tips, palette= 'rainbow')
```
